```python
annual_data["Sales"]=annual_data["Quantity Ordered"]*annual_data["Price Each"]
annual_data.dropna(how="all")
annual_data.head()
```

```python
annual_data.dropna(inplace=True)
annual_data.head()
```

### Create a Column City

```python
### 1st Way of split using apply and lambda 
###annual_data["City"]=annual_data["Purchase Address"].apply(lambda x: x.split(',')[1])

### 2nd Way of split  using apply and lambda
def get_city(address):
    return address.split(',')[1]

annual_data["City"]=annual_data["Purchase Address"].apply(lambda x: get_city(x))

###3rd Way of split using apply and lambda 
def get_state(address):
    return address.split(',')[2].split(' ')[1]

annual_data["State"]=annual_data["Purchase Address"].apply(lambda x: '(' + get_state(x) + ')')
annual_data["Addr Abbv"]=annual_data["Purchase Address"].apply(lambda x: get_city(x) + ' ' + get_state(x))
annual_data.head()
```

### Dropping a column 

```python
annual_data.drop('Addr Abbv', axis=1, inplace=True) ###Removing 1 column
annual_data.head()

### annual_data.drop(['A','B','C'.....],axis=1,inplace=True) ### Removing mroe than 1 column
```

### Q4) What Products are most often sold together 

### For this we'll use Order ID, same Order ID means same product, so we'll see duplicate Order IDs  

```python
df=annual_data[annual_data["Order ID"].duplicated(keep=False)] ### We've 3 options First, last & False and their functions can be seen by pressing (Shift+Tab)
```

```python
df['Grouped']=df.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))
df.head(14)
```

### Dropping Duplicate Values

```python
df=df[["Order ID","Grouped"]].drop_duplicates()
df.head()
```

### Counting Pair of Products 

```python
from itertools import combinations 
from collections import Counter

count=Counter()
for row in df["Grouped"]:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))

count.most_common(10)
```

### Ans:- As we can see its iPhone and Lightning Charging Cable that is most commonly sold together
